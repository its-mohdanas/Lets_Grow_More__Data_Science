## Task-1: Iris Flowers Classification ML Project:

### The iris flowers dataset contains numeric attributes, and it is perfect for beginners to learn about supervised ML algorithms, mainly how to load and handle data. Also, since this is a small dataset, it can easily fit in memory without requiring special transformations or scaling capabilities.

____

## Step 1: Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.simplefilter("ignore")
```

## Step 2: Importing the dataset

```python
name = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
data = pd.read_csv("iris.data", names=name)
X = data.iloc[:,:4].values
y = data.iloc[:,-1]
```

## Step 3: Doing some EDA

```python
#Structure of data
data.shape
```

```python
data.head()
```

```python
#Description of data
data.describe()
```

```python
#Frequency Table
data["class"].value_counts()
```

```python
sb.catplot(x = 'class', hue = 'class', kind = 'count', data = data)
```

```python
sb.set_style("whitegrid")
sb.pairplot(data,hue="class",size=3);
plt.show()
```

## Step 4: Dividing Data Set into train and test (80% + 20%)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

# 

## Step 5: Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

## ->Prediction and Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
y_pred = reg.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## ->Accuracy

```python
accuracy_score(y_test, y_pred)
```

# As a result, the model is 96.70% percent accurate.
